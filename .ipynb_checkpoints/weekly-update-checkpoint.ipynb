```python

def calculate_returns(frame):
#     print(frame)
    start = frame.head(1)
    end = frame.tail(1)
#     print(start)
#     print(end)
    returns = []
    for i in start:
        percent_return = ((end[i][0]-start[i][0])/float(start[i][0]))*100
        returns.append(percent_return)
#     print(returns)
    mean_ret = np.mean(returns)
    return mean_ret

cnt = 0
print(weeks[25], weeks[29])
data = yf.download(string_formatted_ticker_list, start=weeks[15][0].strftime("%Y-%m-%d"), end=weeks[24][1].strftime("%Y-%m-%d"))
print(calculate_returns(data))

for week in weeks[25:29]:
    data = yf.download(string_formatted_ticker_list, start=week[0].strftime("%Y-%m-%d"), end=week[1].strftime("%Y-%m-%d"))
    cnt += 1
    print(cnt)
    print(calculate_returns(data['Close']))
#     print(data)
```
