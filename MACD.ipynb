```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy


# Note in MACD we use 26th prev day from current, so first 26 values from ans are not correct
def MACD(stock_tickers=['AMZN']):


    string_formatted_ticker_list = " ".join(stock_tickers)

    data = yf.download(string_formatted_ticker_list, start="2021-01-01", end="2021-07-02")

#     figure(figsize=(12, 12), dpi=80)


    exp1 = data.Close.ewm(span=12, adjust=False).mean()
    exp2 = data.Close.ewm(span=26, adjust=False).mean()
    macd = exp1-exp2
    exp3 = macd.ewm(span=9, adjust=False).mean()
#     plt.plot(df.index, macd, label='AMD MACD', color = '#EBD2BE')
#     plt.plot(df.index, exp3, label='Signal Line', color='#E5A4CB')
#     plt.legend(loc='upper left')
#     plt.show()
    
    MACD = macd.to_numpy(dtype=None, copy=False)
    EXP3 = exp3.to_numpy(dtype=None, copy=False)
    DATES = data.index.to_numpy(dtype=None, copy=False)
    
    crossover_dates_bullish = []
    crossover_dates_bearish = []
    
    for i in range(1, len(DATES)):
        if MACD[i-1] < EXP3[i-1] and MACD[i] > EXP3[i]:
            crossover_dates_bullish.append(DATES[i])
        if MACD[i-1] > EXP3[i-1] and MACD[i] < EXP3[i]:
            crossover_dates_bearish.append(DATES[i])
#     print("BULLISH", crossover_dates_bullish)
#     print("BEARISH", crossover_dates_bearish)

    return crossover_dates_bullish[-1], crossover_dates_bearish[-1]
    
    
```
